```python
# Imported the required libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loaded the dataset and displayed the first 5 rows.
df = pd.read_csv("titanic.csv")
df.head(5)
```

```python
df.describe()
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Histograms
df[num_cols].hist(figsize=(12,6), bins=30)
plt.suptitle("Histograms of Numeric Features")
plt.show()

# Boxplots
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pairplot (can be heavy on large datasets)
sns.pairplot(df[num_cols])
plt.show()
```

Basic feature-level inferences


Age: Majority between 2040, but children and elderly exist = could influence survival.

Fare: Strongly skewed; higher fares  often 1st class = better survival chances.

Pclass: Negatively correlated with Fare (1st class = expensive).

Embarked: Different survival rates by embarkation point (C > Q > S).

Sex: One of the most important predictors  females had much higher survival.
